4. Accessing List Elements


```python
a = [10, 20, 30, 40, 50]

# Indexing
print(a[0])    # 10
print(a[-1])   # 50 (last element)

# Slicing
print(a[1:4])  # [20, 30, 40]
print(a[:3])   # [10, 20, 30]
print(a[::-1]) # reverse list
```

5. Modifying Lists

```python
a = [10, 20, 30]

a.append(40)           # add one element at end
a.extend([50, 60])     # add multiple elements
a.insert(1, 15)        # insert at specific position
a[2] = 99              # update value
#a.remove(20)           # remove by value
popped = a.pop()       # remove last element
del a[0]               # delete element by index
a.clear()              # remove all elements
```

## Common List Methods

```python
# append()  Add single item at the end
nums = [1, 2, 3]
nums.append(4)
print(nums)  # [1, 2, 3, 4]
```

```python
# extend()  Add multiple items
nums.extend([5, 6])
print(nums)  # [1, 2, 3, 4, 5, 6]
```

```python
# insert()  Insert at specific position

nums.insert(2, 99)
print(nums)  # [1, 2, 99, 3, 4, 5, 6]
```

```python
# remove()  Remove by value

nums.remove(99)
print(nums)
```

```python
#pop()  Remove and return last or specific index
nums.pop()      # removes last element
nums.pop(0)     # removes element at index 0
```

```python
# clear()  Empty the list

nums.clear()
```

```python
# reverse()  Reverse in-place

a = [1, 2, 3]
a.reverse()
print(a)  # [3, 2, 1]
```

```python
# sort()  Sort the list

a = [3, 1, 4, 2]
a.sort()               # ascending
a.sort(reverse=True)   # descending
```

```python
# copy()  Shallow copy
a = [1, 2, 3]
b = a.copy()
print(b)

```

## Nested Lists

```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
print(matrix[1][2])  # 6

# Flattening
flat = [num for row in matrix for num in row]
print(flat)  # [1,2,3,4,5,6,7,8,9]
```

## **List Comprehensions (Pythonic Way)**

```python
nums = [1, 2, 3, 4, 5]
squares = [x**2 for x in nums]
evens = [x for x in nums if x % 2 == 0]
uppercase = [ch.upper() for ch in ['a', 'b', 'c']]

print(squares)  # [1, 4, 9, 16, 25]
print(evens)    # [2, 4]
print(uppercase)  # ['A', 'B', 'C']
```

## **Real-World Use Cases (Industry Level)**

```python
#Data Processing
prices = [100, 200, 300]
with_tax = [p * 1.19 for p in prices]
print(prices)

print(with_tax)
```

```python
#Filtering / Cleaning

data = ["", "Alice", None, "Bob"]
cleaned = [x for x in data if x]  # remove empty / None

print(cleaned)
```

```python
# Merging Multiple Lists

a = [1, 2]; b = [3, 4]
merged = a + b  # [1, 2, 3, 4]

print(merged)
```

```python
# Matrix Operations

matrix = [[1, 2], [3, 4]]
transpose = [[row[i] for row in matrix] for i in range(len(matrix[0]))]
print(transpose)  # [[1, 3], [2, 4]]
```

```python
# Enumerate for Index + Value

fruits = ["apple", "banana", "cherry"]
for i, fruit in enumerate(fruits, start=1):
    print(i, fruit)
```

## **Shallow vs Deep Copy (Important in Industry)**

```python
import copy

a = [[1, 2], [3, 4]]

b = a.copy()             # Shallow copy  linked inner lists
c = copy.deepcopy(a)     # Deep copy  independent copy

a[0][0] = 99
print(b)  # [[99, 2], [3, 4]]
print(c)  # [[1, 2], [3, 4]]
```


Real-World Tip (Industry Use)

Use .copy() for flat lists (like [1, 2, 3]).

Use copy.deepcopy() for nested or complex structures (like lists of lists, dicts of dicts, etc.), to avoid shared-reference bugs.
